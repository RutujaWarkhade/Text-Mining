```python
import pandas as pd
```

```python
df = pd.read_csv("C:/8-text minning/text_mining\homeprices.csv")
df
```

```python
#using pandas to create dummy variables
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(['town'],axis="columns")
final
```

```python
#(interview) what is dummy variable trap?
"""
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you 
can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this 
situation linear regression won't work as expected. Hence you need to drop one column.
NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should
make a habit of taking care of dummy variable trap ou are using is not handling this for you
"""
```

```python
"""
sklearn library takes care of dummy variable   



"""
```

```python
final = final.drop(['west windsor'], axis = "columns")
final
```

```python
X = final.drop('price', axis = "columns")
X
# capital X for input the dataframe
```

```python
y = final.price
#y for targeted column
```

```python
#define the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
#training the model
model.fit(X,y)
```

```python
model.predict(X)
#2600 sqr ft home in new jersey
```

```python
model.score(X,y)
```

```python
model.predict([[34800,0,0]]) #3400 sqr ft home in west windsor
```

```python
model.predict([[2800,0,1]]) #2800sqr ft home in west windsor
```

```python
#first step is to use label encoder to convert town names into numbers
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
```

```python
X
```

```python
y = dfle.price.values
y
```

```python
#now use one hot encoder to create dummy variables for each of the town
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])],remainder = 'passthrough')
#The code creates a ColumnTransformer that one-hot encodes
#the first column #(index 0) of the dataset and leaves
#the remaining columns unchanged (passthrough)
#this is useful when you have a mix of categorical and numerical features 
#in your dataset and want to apply specific transformations
#only to certain columns while keeping others as they are.
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]#droping first column using slicing
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in west robbinsiville
```
